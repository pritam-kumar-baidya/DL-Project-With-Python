```python
import math 
import numpy as np # Linear algebra
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def TanhActivationsFunctions(number):
    tanhstore=[]
    for i in number:
        tanhstore.append(math.tanh(i))
    return tanhstore
```

```python
random_number = np.arange(-20.0,20.0,0.2)
tanh=TanhActivationsFunctions(random_number)
plt.plot(random_number,tanh)
ax=plt.gca()
plt.show()
```
