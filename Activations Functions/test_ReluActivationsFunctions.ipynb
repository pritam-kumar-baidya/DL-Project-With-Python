```python
import math 
import numpy as np # Linear algebra
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def ReluactivationFunction(number):
    store=[]
    for i in number:
        initialvalue=0
        if i>0:
            initialvalue=i
        store.append(initialvalue)
    return store
```

```python
random_number=np.arange(-20.0,20.0,0.2)
relu=ReluactivationFunction(random_number)
plt.plot(random_number,relu)
ax=plt.gca()
plt.show()
```
